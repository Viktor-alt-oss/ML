# Графики сравнения - pairplot() и jointplot()

## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Данные

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

## jointplot

```python
sns.jointplot(x='math score',y='reading score',data=df)
```

```python
sns.jointplot(x='math score',y='reading score',data=df,kind='hex')
```

```python
sns.jointplot(x='math score',y='reading score',data=df,kind='kde')
```

## pairplot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='gender',palette='viridis')
```

```python
sns.pairplot(df,hue='gender',palette='viridis',corner=True)
```

```python
sns.pairplot(df,hue='gender',palette='viridis',diag_kind='hist')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
sns.jointplot(data=df, x='math score', y='reading score', kind='scatter', alpha=0.2 )
```

```python
sns.jointplot(data=df, x='math score', y='reading score', kind='hist')
```

```python
sns.jointplot(data=df, x='math score', y='reading score', kind='kde')
```

```python
sns.jointplot(data=df, x='math score', y='reading score', kind='kde', fill=True)
```

```python
sns.jointplot(data=df, x='math score', y='reading score', hue='gender')
```

```python
sns.pairplot(data=df, hue='gender', corner=True)
```
