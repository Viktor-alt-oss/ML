# plt.subplots()

**ЗАМЕЧАНИЕ: Убедитесь, что Вы запускаете все команды в одной и той же ячейке так же, как мы делаем в этом блокноте и в видео-лекциях!**

Метод plt.subplots() позволяет автоматически разместить несколько наборов осей рядом друг с другом.

Обратите внимание, что здесь мы используем распаковку кортежа (tuple unpacking) для того, чтобы получить сразу два объекта - объект Figure и массив numpy с набором осей:

```python
# Очень похоже на plt.figure(), но здесь мы используем 
# распаковку кортежа для получаения значений fig и axes
fig, axes = plt.subplots()

# Далее с помощью объекта axes указываем значения для атрибутов
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title'); # вызов этого метода с точкой запятой в конце позволяет не отображать Out[]
```

## Рисуем графики на объекте axes 

Как и раньше, мы можем рисовать графики с помощью метода .plot() для каждого из наборов осей, а также использовать методы .set_ для атрибутов.

Давайте посмотрим. Повторимся, что все команды должны быть в одной и той же ячейке:

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

for axe in axes:
    axe.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(a,b)
axes[1].plot(x,y)
```

```python
# Обратите внимание! На выходе получаем двумерный массив
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
```

Частая проблема в matplotlib - это наложение графиков друг на друга. Чтобы решить эту проблему, можно использовать методы **fig.tight_layout()** или **plt.tight_layout()** для автоматической подгонки осей, чтобы не было пересечений:

```python
# Обратите внимание! Здесь мы получаем двумерный массив
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(a,b)
axes[1][1].plot(x,y)  

plt.tight_layout()
```

## Параметры subplots()

Вспомним, что у нас есть два объекта - Figure и axes. Мы можем редактировать свойства этих объектов.

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(12,8))

# СНАЧАЛА УСТАНАВЛИВАЕМ ПАРАМЕТРЫ AXES

# Параметры на уровне объекта axes 
axes[0][0].plot(a,b)
axes[0][0].set_title('0 0 Title')


axes[1][1].plot(x,y)
axes[1][1].set_title('1 1 Title')
axes[1][1].set_xlabel('1 1 X Label')

axes[0][1].plot(y,x)
axes[1][0].plot(b,a)

# ДАЛЕЕ УСТАНАВЛИВАЕМ ОБЩИЕ ПАРАМЕТРЫ FIGURE

# Параметры на уровне объекта Figure
fig.suptitle("Figure Level",fontsize=16)


plt.show()
```

### Ручное указание размеров в subplots()

.subplots_adjust позволяет указывать размеры вручную.

Более полное описание можно почитать здесь: https://matplotlib.org/3.2.2/api/_as_gen/matplotlib.pyplot.subplots_adjust.html

Примеры из этой статьи:

* left = 0.125  # отступ слева
* right = 0.9   # отступ справа
* bottom = 0.1  # отступ снизу
* top = 0.9     # отступ сверху
* wspace = 0.2  # ширина между отдельными наборами осей, измеряемая как доля от средней ширины осей
* hspace = 0.2  # высота между наборами осей, измеряемая как доля от средней высоты осей

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(12,8))

# Параметры на уровне axes 
axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
axes[0][1].plot(y,x)
axes[1][0].plot(b,a)

# С помощью left,right,top, bottom можно двигать наборы осей
# С помощью wspace,hspace можно регулировать расстояния между наборами осей
fig.subplots_adjust(left=None,
    bottom=None,
    right=None,
    top=None,
    wspace=0.9,
    hspace=0.1,)

plt.show()
```

## Экспортируем plt.subplots()

```python
# Обратите внимание! Здесь мы получаем двумерный массив
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(12,8))

axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
axes[0][1].plot(y,x)
axes[1][0].plot(b,a)

fig.savefig('subplots.png',bbox_inches='tight')

plt.show()
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1)

for axe in axes:
    axe.plot(x, y)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x, y)
axes[0][1].plot(x, y)
axes[1][0].plot(x, y)
axes[1][1].plot(a, b)

fig.subplots_adjust(wspace=1, hspace=1)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, dip=200, figsize=())

axes[0][0].plot(x, y)
axes[0][1].plot(x, y)

axes[1][0].plot(x, y)
axes[1][0].set_ylabel('dsfdf')

axes[1][1].plot(a, b)
axes[1][1].set_title('dfsdfsdf')

fig.suptitle('Figure level Title')
plt.tight_layout()

fig.savefig('new_subplots.png', bbox_inches='tight')
```
